```python
array_zeros_1 = numpy.zeros(10)
array_zeros_2 = numpy.zeros((10,5))
array_zeros_3 = numpy.zeros((5,10,10))

print(array_zeros_1,array_zeros_1.shape)
print("-----------------")
print(array_zeros_2,array_zeros_2.shape)
print("-----------------")
print(array_zeros_3,array_zeros_3.shape)
```

```python
array_ones_1 = numpy.ones(10)
array_ones_2 = numpy.ones((10,5))
array_ones_3 = numpy.ones((5,10,10))

print(array_ones_1,array_ones_1.shape)
print("-----------------")
print(array_ones_2,array_ones_2.shape)
print("-----------------")
print(array_ones_3,array_ones_3.shape)
```

```python
array_empty_1 = numpy.empty(10)
array_empty_2 = numpy.empty((10,5))
array_empty_3 = numpy.empty((5,10,10))

print(array_empty_1,array_empty_1.shape)
print("-----------------")
print(array_empty_2,array_empty_2.shape)
print("-----------------")
print(array_empty_3,array_empty_3.shape)
```
