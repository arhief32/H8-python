```python
#1
array1 = np.arange(0,1,0.05)
print(array1, array1.shape)

#2
array1.reshape(10,2)
print(array1)

#3
array2 = array1.dot(2)
print(array2)
print(array2.shape)

#4
matmul_array2 = np.matmul(array1, array2.T)
print(matmul_array2)
print(matmul_array2.shape)

#5
matmul_mean = matmul_array2.mean()
print(matmul_mean)
matmul_median = np.median(matmul_array2, )
print(matmul_median)
matmul_std = matmul_array2.std()
print(matmul_std)
matmul_var = matmul_array2.var()
print(matmul_var)
```
