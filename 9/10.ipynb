# Uniform Distribution

```python
n = 10
start = 10
width = 20
rv_uniform = uniform.rvs(size=n, loc=start, scale=width)
rv_uniform
```

```python
sns.distplot(
    rv_uniform,
    bins=100,
    kde=True,
    color='skyblue',
    hist_kws={"linewidth": 15, "alpha": 1}
)
plt.xlabel = ''
plt.ylabel = ''
plt.show()
```
